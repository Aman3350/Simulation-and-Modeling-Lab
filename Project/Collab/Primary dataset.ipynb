```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import re
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Research data/Primary data.csv')
print(df)
```

```python
def clean_text(text):
    text = str(text)
    text = re.sub(r"http\S+", "", text)   # remove URLs
    text = re.sub(r"@\w+", "", text)      # remove mentions
    text = re.sub(r"#\w+", "", text)      # remove hashtags
    text = re.sub(r"[^\w\s]", "", text)   # remove punctuation
    return text.lower()

# Rename the first column to 'text' if it's not already named 'text'
if 'text' not in df.columns and len(df.columns) == 1:
    df.rename(columns={df.columns[0]: 'text'}, inplace=True)

df['clean_text'] = df['text'].apply(clean_text)
```

```python
analyzer = SentimentIntensityAnalyzer()

def get_polarity_score(text):
    score = analyzer.polarity_scores(text)['compound']  # -1 to +1
    scaled = int((score + 1) * 7)  # convert to 0–14 scale
    return scaled

df['polarity_score'] = df['clean_text'].apply(get_polarity_score)
```

```python
def get_sentiment(score):
    if score > 7:
        return 1   # Positive
    elif score < 7:
        return 2   # Negative
    else:
        return 0   # Neutral

df['sentiment'] = df['polarity_score'].apply(get_sentiment)
```
